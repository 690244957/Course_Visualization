```python
df_cost= pd.read_table("cost.csv",   
                       encoding = 'utf8',
                      sep=',')
                              
df_cost
```

```python
df_GDP=pd.read_table('GDP.csv',
                     encoding='utf8',
                     sep=',')
df_GDP
```

```python
def prepare(df_in, l_variable, l_value):

    df = df_in.copy()
    df.columns = [ int(x.replace('年','')) if x!='地区' else x for x in df.columns]
    years = [ x for x in df.columns if x!='地区'] 
    df_out = pd.melt(df, id_vars=['地区'], value_vars=years)
    df_out.columns = ['地区', l_variable, l_value]
    df_out = df_out.set_index(['地区',  l_variable])
    return (df_out)
```

```python
prepare(df_cost,'年','cost')
```

```python
prepare(df_GDP, "年", "GDP")
```

```python
df_p = prepare(df_GDP, "年", "GDP").join( prepare(df_cost, "年", "cost"), 
                                          how = "left" )
df_p 
```

```python
df_p.query("年==2015")
```

```python
import numpy as np
x = df_p.query("年==2015")['GDP']
y = df_p.query("年==2015")['cost']
fit = np.polyfit(x=x , y=y, deg=1)

ax = df_p.query("年==2015").plot(kind='scatter', x='GDP', y='cost', 
          title = 'GDP与居民消费水平的关系：分省数据')
ax.plot(x, fit[0] * x + fit[1], color='red')
ax.scatter(x, y)
```
